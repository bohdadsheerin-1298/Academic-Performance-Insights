```python
import pandas as pd
df=pd.read_csv("student_marks_dataset.csv")
df.head()

#Basic Cleaning
null_values=df.isnull().sum()
print(null_values)

df["Percentage"]=df["Marks"]/df["MaxMarks"]*100
print(df["Percentage"])

#Class Statistics
class_avg=df["Marks"].mean()
subject_avg=df.groupby(["Subject"])["Marks"].mean()

#Highest and Lowest Scorer
highest=df.loc[df["Marks"].idxmax()]
lowest=df.loc[df["Marks"].idxmin()]

#Pass/Fail Column
df["Result"]=df["Percentage"].apply(lambda x: 'Pass' if x >= 35 else 'Fail')
print(df["Result"])

#Pass Percentage
pass_percent=(df[df["Result"] == 'Pass'].shape[0] / df.shape[0])*100

#subjectwise Topper
subject_topper=df.loc[df.groupby("Subject")["Marks"].idxmax()]

df.to_csv("Cleaned_student_MDS.csv")
```
